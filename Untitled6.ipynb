```
import numpy as np
import matplotlib.pyplot as plt
#데이터 읽기
data    = np.genfromtxt("data.txt", delimiter=',')
x       = data[:, [0]]
y       = data[:, 1]
label   = data[:, 2]
#과제 5-1
x_label0    = x[label == 0]
x_label1    = x[label == 1]
y_label0    = y[label == 0]
y_label1    = y[label == 1]

plt.figure(figsize=(8, 8))
plt.scatter(x_label0, y_label0, alpha=0.3, c='b')
plt.scatter(x_label1, y_label1, alpha=0.3, c='r')
plt.scatter(x, y, alpha=0.3)
plt.show

#과제 5-2

#데이터 값 몇개인가 구하기
l=len(x)  
print(l)

#σ(z)값 계산하는 식
def sigmoid(a):
    return 1.0/(1 + np.exp(-a))

#gradient descent,theta0=w0,theta1=w1,theta2=w2
def GD(x,y,label):
  i = 0
  w0=0.5
  w1=0.4
  w2=0.3
#logistic regression을 gradient descent에 넣기
  while(i<100):
   z=w0+w1*x[i]+w2*y[i]
   w0=w0-(0.2/l)*(sigmoid(z)-label[i])
   w1=w1-(0.2/l)*(sigmoid(z)-label[i])*x[i]
   w2=w2-(0.2/l)*(sigmoid(z)-label[i])*y[i]
   plt.scatter(i,w0,c='r')
   plt.scatter(i,w1,c='g')
   plt.scatter(i,w2,c='b')
   i=i+1
#과제 5-2 plot  
plt.figure(figsize=(8, 8))
GD(x,y,label)
plt.show

```
